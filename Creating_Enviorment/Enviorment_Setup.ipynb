After that, you can use commands related to `conda`.

---

## Creating a New Virtual Environment

Creating a new virtual environment for each project is a good practice. It helps to keep dependencies separate and avoid conflicts.

To create a new virtual environment, use the following command:

```bash
conda create -p venv python=3.10
```

### Notes:
- **`venv`**: This is the name of the virtual environment. You can replace it with any name you prefer.
- **Python Version**: Specifying a Python version is optional. If you donâ€™t specify a version (e.g., `python=3.7`), it will use the default Python version.

---

## Activating the Virtual Environment

To activate the virtual environment, use the command:

```bash
conda activate "C:\Users\AYAN HALDAR\Documents\MY_MACHINE_LEARNING_JOURNEY\BASICS\venv"
```

---

## Installing Required Packages

You can install required packages individually using:

```bash
pip install numpy pandas matplotlib scikit-learn
```

### Alternative: Using a `requirements.txt` File

Instead of installing packages one by one, you can create a `requirements.txt` file containing the required package names and their versions. Then, install all packages at once using:

```bash
pip install -r requirements.txt
```

---

## Example `requirements.txt` File

```plaintext
numpy
pandas
matplotlib
scikit-learn
```

This ensures a structured and efficient way to manage your project dependencies.

## Other Alternative Way to Create a Virtual Environment

We can also create a virtual environment by using the following command:

```bash
python -m venv myenv
```

- Here, **venv** is the name of the virtual environment. You can give any name to the virtual environment. This will create a virtual environment with the default version of Python.

To activate the above environment:

```bash

myenv\Scripts\activate
```
