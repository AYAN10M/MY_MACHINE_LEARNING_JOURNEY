# 0.An end-to-end Scikit-Learn workflow

Our objective is to find the target column from the heart disease csv file.


1.   First we will drop target our target column from X to make it our feature matrix and another y which will be out output matrix.
2.   First of all this is a "classification" type problem.




```python
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv("csv-files/heart-disease.csv")
```

```python
heart_disease
```

```python
# Step 1: Creating features and label dataframes

# X is features Matrix
# Consider features Matrix as input data
X = heart_disease.drop("target", axis = 1)

# Y is label
# Consider this as output
y = heart_disease["target"]
```

```python
# Step 2: Choosing the rigth model and hyperparameter

# Hyperparameters are defined as the parameters that are explicitly defined by the user to control the learning process for better result

from sklearn.ensemble import RandomForestClassifier

# clf stands for "classifier". Alternatively we can use the term "model" too.

clf = RandomForestClassifier()

# For now we will use the default hyperparameters
clf.get_params()
```

```python
# Step 3: Fit the model to the trainning data
# We need to split the data in training and test data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
clf.fit(X_train, y_train);
```

```python
y_preds = clf.predict(X_test)
y_preds
```

```python
y_test
```

```python
# Step 4: Evaluate the model
# score: Returns the mean accuracy on the given test data and labels
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

This shows that while performing operatios on trained data it will give result with 100% accuracy and if the operations were performed in random data it will give the result in 75% accuracy.

```python
from sklearn.metrics import *
```

```python
print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
# Step 5: Improve a model
# Try different types of n_estimators

np.random.seed(42)

for i in range(10, 100, 10):
  print(f"Trying model with {i} estimators...")
  clf = RandomForestClassifier(n_estimators = i).fit(X_train, y_train)
  print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
  clf.fit(X_train, y_train)
  print("")
```

```python
# Step 6: Save a model and load it
import pickle

pickle.dump(clf, open("random_forest_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
